**Student Case-Study**

```
import io
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
from google.colab import files
uploaded = files.upload()
```

```
student_data=pd.read_csv(io.BytesIO(uploaded['student_data.csv']))
```

```
student_data["study_time"].value_counts()
```

```
sns.relplot(x="absences",y="G3",data=student_data,col="study_time")
plt.show()
```

```
sns.relplot(x='G1' ,y='G3', data=student_data, kind ='scatter',col="schoolsup",
            col_order=["no","yes"],row="famsup",row_order=["yes","no"])
plt.show()
```
